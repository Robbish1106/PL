```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
BANK = pd.read_csv("https://raw.githubusercontent.com/Robbish1106/PL/main/hw1/A17000000J-020024-sW1.csv", encoding='utf-8') #提供各種繳付國民保險金服務的銀行
BANK.head()
```

```python
first = BANK.drop(columns=['金資代號','代收國保費','轉帳代繳國保費','網路ATM繳納國保費','行動APP繳納國保費']) #提出"網路銀行繳納國保費"項
first.tail()
```

```python
NB = first.drop(first[first['網路銀行繳納國保費']=='無'].index) #把無的排掉
NB
```

```python
second = BANK.drop(columns=['金資代號','代收國保費','轉帳代繳國保費','網路銀行繳納國保費','行動APP繳納國保費']) #提出"網路ATM繳納國保費"項
second.tail()
```

```python
NA = second.drop(second[second['網路ATM繳納國保費']=='無'].index) #把無的排掉
NA
```

```python
p1 = list(NB['金融機構名稱'])
p2 = list(NA['金融機構名稱'])
```

```python
ps1 = {str(i) for i in p1}
ps2 = {str(i) for i in p2}
```

```python
print(ps1|ps2) #聯集:有"網路ATM繳納國保費"和"網路銀行繳納國保費"兩項服務或其一的銀行
```

```python
print(ps1&ps2) #交集:"網路ATM繳納國保費"和"網路銀行繳納國保費"兩項服務皆有的銀行
```

```python
print(ps1-ps2) #差集:只有"網路銀行繳納國保費"一項服務的銀行
```

```python
print(ps1^ps2) #對稱差集:只有"網路銀行繳納國保費"一項或"網路ATM繳納國保費"一項服務的銀行
```
