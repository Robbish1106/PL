```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PL1 = pd.read_csv("https://raw.githubusercontent.com/Robbish1106/PL/main/hw1/STOCK_DAY_2330_202109.csv", encoding='utf-8')
PL2_f = pd.read_csv("https://raw.githubusercontent.com/Robbish1106/PL/main/hw1/STOCK_DAY_2330_202209.csv", encoding='utf-8')
PL2 = PL2_f.drop([20])
```

```python
p1 = list(PL1['收盤價'])
p1
```

```python
p2 = list(PL2['收盤價'])
p2
```

```python
ps1 = {float(i) for i in p1}
ps2 = {float(i) for i in p2}
```

```python
print(ps1|ps2) #聯集
```

```python
print(ps1&ps2) #交集
```

```python
print(ps1-ps2) #差集
```

```python
print(ps1^ps2) #對稱差集
```
