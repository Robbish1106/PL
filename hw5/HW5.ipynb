```python
#爬蟲
import urllib.request as req
import json
import pandas as pd
import numpy as np

url= "https://www.ptt.cc/bbs/C_Chat/M.1651039300.A.D0D.html"
request=req.Request(url, headers={
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
})
with req.urlopen(request) as response:
    data=response.read().decode("utf-8")
import bs4
root=bs4.BeautifulSoup(data, "html.parser")
articles = root.find_all('div', 'push')
for article in articles:
    messages = article.find('span', 'f3 push-content').getText()
temp = []
for article in articles:
    messages = article.find('span', 'f3 push-content').getText()
    temp.append(messages)
```

```python
import string
string.punctuation
http = ['h']
```

```python
new_temp = []
for word in temp:
    for letter in word:
        if letter in string.punctuation:
            word = word.replace(letter,"")
        elif letter in http:
            word = word.replace(word,"")
    new_temp.append(word)
```

```python
try:
    while True:
        new_temp.remove("")
except ValueError:
    pass
print(new_temp)
```

```python
data = pd.DataFrame(new_temp)
data
```

```python
#文字分析
import jieba
import jieba.analyse
import pandas as pd
import numpy as np
from collections import Counter
```

```python
terms=[]
stopword=[' ','[',']',"在","都","誰","也","但","來","個","再","的","和","是","有","更","會","可能","有何","從","對","就", '\n','越','為','這種','多','越來',"請問","請","了","嗎","報","無",'／','《','》','，','。','「','」','(',')','！','？','、','…',':',"[","]","~",'"','&','/','（']
```

```python
for i in range(len(data[0])):
  for j in list(jieba.cut(data[0][i],cut_all=False)):
    if j not in stopword:
      terms.append(j)
```

```python
terms
```

```python
#出現數
results = pd.value_counts(terms)
results
```

```python
#拆解字詞dict
firstcolum = {}          
relationships = {}  
toprow = []  
for i in range(len(data[0])):
        poss = jieba.cut(data[0][i], cut_all = False)
        toprow.append([])
        for w in poss:
            if w not in stopword:
                toprow[-1].append(w)        
            if firstcolum.get(w) is None and w not in stopword:
                relationships[w] = {}            
```

```python
term_dic = dict()
for sentence in toprow:
    for term in sentence:
        if term not in term_dic:
            term_dic[term] = {}
```

```python
for i in range(len(toprow)):
    for term1 in term_dic:
        num = 0
        for term2 in toprow[i]:
            if term1 == term2:
                num += 1
        term_dic[term1][data[0][i]] = num
```

```python
matrix = pd.DataFrame.from_dict(term_dic)
matrix
```

```python
#共現矩陣
coocc = matrix.T.dot(matrix)
coocc
```
